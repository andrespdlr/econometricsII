```python
import random
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
for i in range(2500):
    x = random.uniform(0, 1)
    y = random.randint(0, 1)
    X.append(x)
    Y.append(y)
```

```python
plt.plot(X, Y, 'o', markersize=.5)
```

```python
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
betas = model.params.tolist()
```

```python
betas
```

```python
for i in range(100):
    x = random.uniform(0, 1000)
    y_hat = betas[0] + (betas[1] * x)
    print(y_hat, end=' ')
```

```python
plt.plot(X, Y, 'o', markersize=.5)
plt.plot([0, 1], [betas[0], betas[0] + (betas[1] * 1)], color='blue', linestyle='solid', linewidth=2)
```

```python
from scipy.stats import norm
```

```python
X = []
Y = []
for i in range(2500):
    x = random.uniform(-100, 100)
    y = norm.cdf(x)
    X.append(x)
    Y.append(y)
```

```python
plt.plot(sorted(X), sorted(Y), '-')
```
